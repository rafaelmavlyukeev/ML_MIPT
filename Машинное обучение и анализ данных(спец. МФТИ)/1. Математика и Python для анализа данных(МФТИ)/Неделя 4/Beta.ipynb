# Центральная предельная теорема своими руками

## Beta распределение

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
```

```python
a = 0.5
b = 0.5
E = a / (a + b)
D = (a * b) / ((a + b)**2 * (a + b +1))

#Сгенерируйте из него выборку объёма 1000
beta = sts.beta(a, b)
sample = beta.rvs(size=1000)
#Постройте гистограмму выборки и нарисуйте поверх неё теоретическую плотность распределения вашей случайной величины.
plt.hist(sample, density =True, bins=20, alpha=0.5, label='hist samples beta')
plt.ylabel('number of samples')
plt.xlabel('$x$')

#теоретическая плотность распределения случайной величины
x = np.linspace(0, 1, 100)
plt.plot(x, beta.pdf(x), 'r-', lw=5, alpha=0.7, label='beta pdf')
plt.legend(loc='best')

plt.show()

# функция построения гистограммы распределений выборочных средних 
# и плотности соответствующего нормального распределения 
# sizeSamples - выбороки объёма n
def betaF(sizeSamples, E, D):
    n = sizeSamples
#генерация выборок
    values = np.array([beta.rvs(size=n) for x in range(1000)])
#вычисление выборочных средних
    meanVal = values.mean(axis = 1)
    plt.hist(meanVal, bins= 30, density =True, alpha=0.75, label='hist mean n ' + str(n))
#зададим нормальное распределение
    mu = E
    sigma = math.sqrt(D/n)
    print('mu=' , mu)
    print('sigma=' , sigma)
    norm_rv = sts.norm(loc=mu, scale=sigma)
    x = np.linspace(0,1,100)
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf, 'r-', lw=3, alpha=0.7, label='beta pdf n ' + str(n))
    plt.ylabel('samples')
    plt.xlabel('$x$')
    plt.legend(loc='best')

    plt.show()
```

```python
betaF(5, E, D)
```

```python
betaF(10, E, D)
```

```python
betaF(50, E, D)
```

Вывод: 
Распределение выборочных средних для функции beta хорошо описывается нормальным распределением уже при n = 5.
С ростом n точность аппроксимации увеличивается.
