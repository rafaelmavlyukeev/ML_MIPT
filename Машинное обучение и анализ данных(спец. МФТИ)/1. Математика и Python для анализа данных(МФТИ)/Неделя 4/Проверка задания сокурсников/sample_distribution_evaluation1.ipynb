```python
import numpy as np
import math
import scipy.stats as sts
from scipy.stats import pareto
import matplotlib.pyplot as plt
%matplotlib inline
```

## Генерация СВ из распределения Парето

```python
b = 5.35 #parameter for Pareto _ параметр b для Парето
sample = pareto.rvs(b, size=1000)
```

## Гистограмма выборки и плотность распределения

```python
#В данном блоке идет построение гистограммы выборки, на которую накладывается функция плотности распределения Парето
x = np.linspace(0.5, 2.5, 100)
plt.figure(figsize = (7, 5))
pdf = pareto.pdf(x, b)
plt.plot(x, pdf,'g-', alpha=0.5, linewidth = 5, color = 'navy')
plt.hist(sample, normed = True, color = 'silver')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

## Функция генерации выборок и подсчета выборочного среднего

```python
# С целью упрощения кода написана функция, которая принимает на вход параметр b распределения Парето и n - объем выборки
# а возвращает список выборочных средних. Выборки генерируются 1000 раз.
def dif_samples(b, n):
    sample_means = []
    for i in range(1000):
        sample = pareto.rvs(b, size = n)
        sample_mean = 0
        for j in sample:
            sample_mean = sample_mean + j/n
        sample_means.append(sample_mean)
    return sample_means
```

## Построение и аппроксимация на выборках разного объема

В последующих блоках задается объем выборки, после чего с помощью вышеописанной функции создается список средних выборочных.
Затем вычисляются статистики - математическое ожидание(среднее, Ex) и дисперсия(разброс, Dx).
После стротся гистограмма выборки, на которую накладывается функция плотности нормального распределения

```python
# Выборка объема n=3
n = 3
N3 = dif_samples(b, n)

Ex = pareto.mean(b)
Dx = pareto.var(b)/n
norm_rv = sts.norm(Ex, math.sqrt(Dx))

x = np.linspace(0.5, 2.5, 100)
plt.figure(figsize = (7, 5))
pdf = norm_rv.pdf(x)
plt.plot(x, pdf,'g-', alpha=0.5, linewidth = 5, color = 'navy')
plt.hist(N3, normed = True, color = 'silver')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
# Выборка объема n=10
n = 10
N10 = dif_samples(b, n)

Ex = pareto.mean(b)
Dx = pareto.var(b)/n
norm_rv = sts.norm(Ex, math.sqrt(Dx))

x = np.linspace(0.5, 2.5, 100)
plt.figure(figsize = (7, 5))
pdf = norm_rv.pdf(x)
plt.plot(x, pdf,'g-', alpha=0.5, linewidth = 5, color = 'navy')
plt.hist(N10, normed = True, color = 'silver')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
#Выборка объема n=70
n = 70
N70 = dif_samples(b, n)

Ex = pareto.mean(b)
Dx = pareto.var(b)/n
norm_rv = sts.norm(Ex, math.sqrt(Dx))

x = np.linspace(0.5, 2.5, 100)
plt.figure(figsize = (7, 5))
pdf = norm_rv.pdf(x)
plt.plot(x, pdf,'g-', alpha=0.5, linewidth = 3, color = 'navy')
plt.hist(N70, normed = True, color = 'silver')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

## Вывод
С увеличением объема выборки n распределение сглаживается и все больше становится похожим на нормальное.
Сооответственно, точность аппроксимации распределения выборочных средних нормальным распределением также улучшается. 
На графиках видно, что при большем объеме выборки n плотность распределения все точнее "ложится" на гистограмму выборки.
