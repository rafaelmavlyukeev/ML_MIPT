```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

Возьмем Гамма-распределение $\Gamma (k=2, \theta=1)$

```python
k = 2
theta = 1
gamma_rv = sts.gamma(k)
```


# 1. Плотность вероятности


```python
#Рисуем теоретическую кривую плотности вероятности
x = np.linspace(0,10,1000)
plt.plot(x, gamma_rv.pdf(x), label='Функция плотности вероятности')

#Рисуем эмпирическую плотность вероятности - нормированную гистограмму
sample = gamma_rv.rvs(1000)
plt.hist(sample, density=True, bins=50, label='Нормированная гистограмма для выборки')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper right')
```


# 2. Оценка распределения выборочного среднего

Будем делать выборки объема n = 5, 10, 50.

Далее везде:

* n - объем одной выборки,
* N = 1000 - количество выборок,
* k = 2 - параметр Гамма-распределения.

Математическое ожидание и дисперсия случайной величины, имеющей гамма-распределение, имеют вид:

$$\mathbb{E}[X] = k{\theta}$$
$$\mathbb{D}[X] = k{\theta^2}$$

Для наших параметров $(k = 2, \theta = 1)$
$$\mathbb{E}[X] = 2$$
$$\mathbb{D}[X] = 2$$

Распределение выборочного вида, согласно ЦПТ будет иметь вид:
$$X \sim N(2, 2/n)$$


**2.1 n = 5**

```python
n = 5
N = 1000
sample = list()

for i in range(N):
    sample.append(gamma_rv.rvs(n).mean())    
```

```python
#Увеличиваем размер графика
plt.rcParams['figure.figsize'] = [9.6, 6.4]

#Рисуем гистограмму
plt.hist(sample, density=True, bins=40, label='Распределение среднего для n = 5')

#Рисуем теоретическую кривую плотности вероятности
# Мат.ожидание для гамма-распределения = k
# Дисперсия для гамма-распределения = k

norm_rv = sts.norm(loc = k, scale = (k/n)**0.5)
x = np.linspace(0,10,1000)
plt.plot(x, norm_rv.pdf(x), label='Функция плотности вероятности')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.xlim(0.5,3.5)
plt.legend(loc='upper right')
```
